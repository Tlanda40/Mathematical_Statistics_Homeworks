```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([.34, 1.38, -.65, .68, 1.40, -.88, -.30, -1.18, .50, -1.75])
y = np.array([.27, 1.34, -.53, .35, 1.28, -.98, -.72, -.81, .64, -1.59])
 
n = len(x)
```

```python
# (a)
b_a = (n*np.sum(x*y) - np.sum(x)*np.sum(y)) / (n*np.sum(x**2) - np.sum(x)**2)
a_a = (np.sum(y) - b_a*np.sum(x)) / n
 
# (b)
d_b = (n*np.sum(x*y) - np.sum(x)*np.sum(y)) / (n*np.sum(y**2) - np.sum(y)**2)
c_b = (np.sum(x) - d_b*np.sum(y)) / n

slope_b    =  1 / d_b
intercept_b = -c_b / d_b
```

```python
x_line = np.linspace(-2, 1.8, 300)

fig, ax = plt.subplots(figsize=(8, 7))
fig.patch.set_facecolor('#0f1117')
ax.set_facecolor('#0f1117')
 
ax.grid(color='#2a2d3a', linewidth=0.6, zorder=0)
ax.axhline(0, color='#444', linewidth=0.8)
ax.axvline(0, color='#444', linewidth=0.8)
 
ax.scatter(x, y, color='#00d4ff', s=70, zorder=5,
           edgecolors='white', linewidths=0.5, label='Data points')
 
ax.plot(x_line, a_a + b_a * x_line,
        color='#ff6b6b', linewidth=2, zorder=4,
        label=f'(a) $y = {a_a:.3f} + {b_a:.3f}x$')
 
ax.plot(x_line, slope_b * x_line + intercept_b,
        color='#ffd166', linewidth=2, linestyle='--', zorder=4,
        label=f'(b) $x = {c_b:.3f} + {d_b:.3f}y$  →  $y = {slope_b:.3f}x {intercept_b:+.3f}$')
 
ax.set_xlim(-2.1, 1.9)
ax.set_ylim(-2.0, 1.7)
ax.set_xlabel('x', color='#ccc', fontsize=13)
ax.set_ylabel('y', color='#ccc', fontsize=13)
ax.set_title('Least Squares Fits: y vs x', color='white', fontsize=15, pad=14)
ax.tick_params(colors='#aaa')
for spine in ax.spines.values():
    spine.set_edgecolor('#333')
 
ax.legend(facecolor='#1a1d2e', edgecolor='#444', labelcolor='white', fontsize=10)
 
plt.tight_layout()
plt.savefig('least_squares_plot.png', dpi=150, bbox_inches='tight',
            facecolor='#0f1117')
plt.show()
 
print(f"(a)  a = {a_a:.4f},  b = {b_a:.4f}")
print(f"(b)  c = {c_b:.4f},  d = {d_b:.4f}")
print(f"(b) as y=f(x):  slope = {slope_b:.4f},  intercept = {intercept_b:.4f}")
```
