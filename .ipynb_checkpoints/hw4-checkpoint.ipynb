```python
import numpy as np
import matplotlib.pyplot as plt
rvs = np.random.uniform(0.0, 1.0, size=1000)
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1)
x = np.arange(1, 1001)
S_n = np.cumsum(rvs)
ax.set_xlim(0, 1005)
ax.set_ylim(0, 600)
ax.set_xlabel('n')
ax.set_ylabel('S_n')
ax.scatter(x, S_n, s = 1.5, alpha = 0.7, color = 'red')
plt.show()
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1)
ax.set_xlim(0, 1000)
ax.set_ylim(0, 1)
ax.set_xlabel('n')
ax.set_ylabel('S_n/n')
ax.scatter(x, S_n/x, s = 1.5, alpha = 0.7, color = 'red')
plt.show()
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1)
ax.set_xlim(0, 1000)
ax.set_ylim(-25, 25)
ax.set_xlabel('n')
ax.set_ylabel('S_n-n/2')
ax.scatter(x, S_n-(x/2), s = 1.5, alpha = 0.7, color = 'red')
plt.show()
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1)
ax.set_xlim(0, 1000)
ax.set_ylim(-0.5, 0.5)
ax.set_xlabel('n')
ax.set_ylabel('(S_n-n/2)/n')
ax.scatter(x, (S_n-(x/2))/x, s = 1.5, alpha = 0.7, color = 'red')
plt.show()
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1)
ax.set_xlim(0, 1000)
ax.set_ylim(-2, 2)
ax.set_xlabel('n')
ax.set_ylabel('(S_n-n/2)/sqrt(n)')
ax.scatter(x, (S_n-(x/2))/np.sqrt(x), s = 1.5, alpha = 0.7, color = 'red')
plt.show()
#IT'S NOT POSSIBLE TO TELL THAT THIS VARIABLE IS NORMALLY DISTRIBUTED ABOUT 0 JUST
#FROM THIS GRAPH. THE GRAPH AFTER THIS ONE, HOWEVER, IS A HISTORGRAM OVER MANY GENERATIONS
#OF THIS VARIABLE, WHICH REVEALS ITS NORMAL DISTRIBUTION!
```

```python
N = 10000
data = np.zeros(N, dtype=float)
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1)
for i in range(N):
    rvs = np.random.uniform(0.0, 1.0, size=1000)
    S_n = np.sum(rvs)
    data[i] = (S_n-(1000/2))/np.sqrt(1000)
ax.hist(data, bins=30)
ax.set_xlim(-2, 2)
ax.set_ylim(0, 1300)
ax.set_xlabel('(S_1000-1000/2)/sqrt(1000)')
ax.set_ylabel('Frequency')
plt.show()
#THIS IS THAT HISTOGRAM.
```
