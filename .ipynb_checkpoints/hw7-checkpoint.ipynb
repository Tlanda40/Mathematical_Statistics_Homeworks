```python
import numpy as np
from scipy import stats

data = np.genfromtxt('gamma-ray.txt', delimiter='\t', skip_header=1)
seconds = data[:, 0]
counts  = data[:, 1].astype(int)
n = len(counts)

print(f"n = {n} observations")
```

```python
# MLE under H0
lambda_hat = counts.sum() / seconds.sum()
mu_i       = lambda_hat * seconds

print(f"Total counts:  {counts.sum()}")
print(f"Total time:    {seconds.sum():.1f} s")
print(f"lambda_hat:    {lambda_hat:.6f} counts/s")

# LRT statistic
lrt_terms = np.where(counts > 0,
                     counts * np.log(counts / mu_i) - counts + mu_i,
                     mu_i)
LRT = 2 * lrt_terms.sum()

# Pearson X^2 statistic
X2 = np.sum((counts - mu_i)**2 / mu_i)

# Degrees of freedom
df = n - 1

# p-values
p_LRT = 1 - stats.chi2.cdf(LRT, df)
p_X2  = 1 - stats.chi2.cdf(X2,  df)

print(f"\nResults:")
print(f"LRT statistic (Lambda): {LRT:.4f}")
print(f"Pearson X^2: {X2:.4f}")
print(f"Degrees of freedom: {df}")
print(f"\nLRT  p-value: {p_LRT:.4f}")
print(f"X^2  p-value: {p_X2:.4f}")
print(f"\nchi2_{df} critical values:")
print(f"alpha=0.05: {stats.chi2.ppf(0.95, df):.2f}")
print(f"alpha=0.01: {stats.chi2.ppf(0.99, df):.2f}")
```
